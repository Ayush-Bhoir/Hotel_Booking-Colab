```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('hotel_bookings.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

**Null Values**

```python
# There was 4 missing values in children so i took mean()
df['children'] = df['children'].fillna(df['children'].mean())
print(df)
```

```python
# 488 Null values in Country
df['country'] = df['country'].fillna('Others')
df
```

```python
# 16340 Null values in Agent
# agent is in numeric column so we can take mean() of column to fill Null values
df['agent'] = df['agent'].fillna(df['agent'].mean())
df
```

```python
# 112593 Null values in company so we can remove this column
df = df.drop('company' , axis = 1)
```

**Replacing 0/1**

```python
df['is_canceled'] = df['is_canceled'].replace([ 1 , 0 ] , [ 'Yes' , 'No' ] )
df
```

```python
print(df['is_canceled'].unique())
```

```python
df['is_repeated_guest'] = df['is_repeated_guest'].replace([ 1 , 0 ] , [ 'True' , 'False' ] )
df
```

```python
print(df['is_repeated_guest'].unique())
```

**Remove Duplicates**

```python
df = df.drop_duplicates()
```

**Datatypes**

```python
df.dtypes
```

```python
# is_canceled we changed 0/1 to Yes/No
df['is_canceled'] = df['is_canceled'].astype('object')
df
```

```python
# children from float -> int
df['children'] = df['children'].astype('int')
df
```

```python
# is_repeated_guest i changed 0/1 to True/False
df['is_repeated_guest'] = df['is_repeated_guest'].astype('object')
df
```

```python
# agent float -> int
df['agent'] = df['agent'].astype('int')
df
```

```python
# Changed datatype from obj -> datetime
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
df
```

**Creating Visuals**

**1) pie chart**

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (12,6))
a = df['hotel'].value_counts()
plt.pie(a ,
        colors=['mediumspringgreen','aquamarine'] ,
        autopct = '%1.1f%%' ,
        labels = a.index ,
        wedgeprops={'width': 0.5}
        )
plt.title('Hotel Type Distribution')
plt.show()
```

**2) Horizontal Bar Chart**

```python
plt.figure(figsize = (12,6))

b = df['market_segment'].value_counts().sort_values(ascending = False)
ax = b.plot(kind='barh',
       color='crimson' ,
       edgecolor = 'black')
plt.title('Distribution of Market Segment ')
plt.ylabel('Market  Segmentation')
plt.xlabel('Counts')
plt.grid(True , alpha = 0.3)

for bar in ax.containers :
  ax.bar_label(bar)

plt.show()
```

**3) Line Chart**

```python
plt.figure(figsize = (12,6))

c = df['arrival_date_month'].value_counts()
plt.plot(c ,
         color = 'lawngreen' ,
         marker = 'o')
plt.title('Monthly Booking Trend')
plt.xlabel('Month')
plt.ylabel('Number of Bookings')
plt.grid(True , alpha = 0.3)
plt.show()
```

**4) Line Chart**

```python
# Leaad time trend
plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
plt.plot(df['lead_time'],
         color = 'green',
         marker = '.',
         )
plt.title('Lead Time Trend on 10-60')
plt.grid(True , alpha = 0.3)
plt.xlabel('Lead Time')
plt.ylabel('Counts')
plt.xlim([10,60])

plt.subplot(1,2,2)
plt.plot(df['lead_time'],
         color = 'green',
         marker = '.',
         )
plt.title('Lead Time Trend on 600-650')
plt.grid(True , alpha = 0.3)
plt.xlabel('Lead Time')
plt.ylabel('Counts')
plt.xlim([600,650])

plt.tight_layout()
plt.show()

```

**5)Pie Chart**

```python
plt.figure(figsize=(12, 6))

c = df['meal'].value_counts()
c.plot(kind='pie',
       autopct='%1.1f%%',
       colors=['gold', 'lightblue', 'lightgreen','red'] ,
       startangle = 90)
plt.title('Meal Type Distribution')
plt.show()
```

**6) Donut Chartt**

```python
plt.figure(figsize=(12, 6))

d = df['is_canceled'].value_counts()
d.plot(kind = 'pie',
            autopct='%1.1f%%',
            colors=['coral', 'skyblue','yellow'],
            wedgeprops={'width':0.5}
            )
plt.title('Reservation Status Cancled or Not')
plt.show()
```

**7) Histogram**

```python
plt.figure(figsize=(10, 5))

e = df['arrival_date_week_number']

plt.hist(e,
         bins=5,
         color='crimson',
         edgecolor = 'black')
plt.title('Distribution of Hotel Booking according to Week number ')
plt.xlabel('Arrival week number')
plt.ylabel('Frequency')
plt.show()
```

**8)Column chart**

```python
plt.figure(figsize=(12, 6))

f = df['arrival_date_year'].value_counts()

f.plot(kind = 'bar' ,
       color = 'c' ,
       edgecolor = 'black' ,
      )
plt.title('Count of Arrival Date per Year')
plt.xticks(rotation = 0)
plt.grid(True , alpha = 0.3)
plt.show()
```

**9) Stacked Bar Chart**

```python
# Stacked Bar Chart
plt.figure(figsize=(12,6))

abd = df.groupby('assigned_room_type')['is_repeated_guest'].value_counts().unstack().fillna(0)

abd.plot(kind = 'bar' ,
         stacked = True ,
         ax = plt.gca() ,
         color = ['navy','mediumblue'])

plt.title('Stacked Bar Chart : Guest is Repeating or Not for Assigned Room Types ')
plt.xlabel('Room types')
plt.ylabel('Count')
plt.xticks(rotation = 0)
plt.grid(True , alpha = 0.3)
plt.legend(title = 'Attrition' , loc = 'upper right')
plt.show()
```

**10) Box Plot**

```python
plt.figure(figsize = (12,8))
df.boxplot(column = 'arrival_date_week_number' , by = 'distribution_channel')
plt.title('Arrival Week number by Channel Distribution ')
plt.suptitle('')
plt.xlabel('Distribution Channel')
plt.ylabel('arrival week number ')
plt.grid(True , alpha = 0.3)
plt.show()
```

**11)Heat Map**

```python
import seaborn as sns

plt.figure(figsize = (12,8))

numericcol = ['lead_time' , 'stays_in_week_nights' , 'stays_in_weekend_nights' , 'arrival_date_day_of_month' , 'arrival_date_week_number' ]

sns.heatmap(df[numericcol].corr() ,
            annot = True ,
            cmap = 'RdYlGn' ,
            linewidths = 0.5)
plt.title('Correlation Heatmap of Key Numeric Columns')
plt.show()
```

**12) Scatter Plot**

```python
plt.figure(figsize=(8, 4))

sns.scatterplot(
    x='total_of_special_requests',
    y='adr',
    data=df,
    color = 'black'
    )
plt.title('Special Requests vs ADR')
plt.show()
```
